```python
#Q1.

import scipy.stats as stats

# Given data
mean = 50
std_dev = 5
confidence_level = 0.95
sample_size = 100  # You need to specify the sample size to determine the degrees of freedom

# Calculate the standard error
standard_error = std_dev / (sample_size**0.5)

# Calculate the margin of error
margin_of_error = stats.norm.ppf(1 - (1 - confidence_level) / 2) * standard_error

# Calculate the confidence interval
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

# Print the results
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
#Q3.

import numpy as np
from scipy.stats import chi2_contingency

# Create the contingency table
observed = np.array([[20, 15],
                     [10, 25],
                     [15, 20]])

# Perform the chi-square test
chi2, p, dof, expected = chi2_contingency(observed)

# Print the results
print("Chi-square statistic:", chi2)
print("P-value:", p)


#Interpreting the results:

#    Chi-Square Statistic: The chi-square statistic is a measure of the difference between the observed and expected frequencies in the contingency table. A higher value indicates a larger discrepancy between the observed and expected values.

#    P-value: The p-value represents the probability of obtaining the observed data (or more extreme data) assuming that there is no association between the variables (i.e., the groups are independent). A lower p-value indicates that the observed data is less likely to occur under the assumption of independence.

#Interpretation of the p-value:

#    If the p-value is small (typically less than 0.05), it suggests that there is a significant association between the variables.
#    If the p-value is large (greater than 0.05), it suggests that the observed association could have occurred by chance, and there might not be a significant relationship between the variables.

#In your case, you would examine the p-value to determine whether there is a significant association between Group A and Group B with respect to the outcomes. If the p-value is small, you might conclude that there is evidence of an association. If the p-value is large, you might conclude that the association could have occurred by chance, and there is no strong evidence of an association.
```

```python
#Q4.


import scipy.stats as stats
import math

# Given data
sample_size = 500
number_of_smokers = 60
confidence_level = 0.95

# Calculate the sample proportion
sample_proportion = number_of_smokers / sample_size

# Calculate the z-score corresponding to the confidence level
z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# Calculate the margin of error
margin_of_error = z_score * math.sqrt((sample_proportion * (1 - sample_proportion)) / sample_size)

# Calculate the confidence interval
confidence_interval_lower = sample_proportion - margin_of_error
confidence_interval_upper = sample_proportion + margin_of_error

# Print the results
print("Sample proportion:", sample_proportion)
print("Z-score:", z_score)
print("Margin of Error:", margin_of_error)
print("95% Confidence Interval:", (confidence_interval_lower, confidence_interval_upper))
```

```python
#Q5.

import scipy.stats as stats

sample_mean = 75
sample_std_dev = 12
confidence_level = 0.90
sample_size = 100  # You need to provide the sample size for accurate calculations

# Calculate the margin of error
margin_of_error = stats.norm.ppf((1 + confidence_level) / 2) * (sample_std_dev / (sample_size ** 0.5))

# Calculate the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print("Confidence Interval:", (lower_bound, upper_bound))

#The calculated 90% confidence interval is a range of values within which we can be 90% confident that the true population mean lies. In this case, with a sample mean of 75 and a standard deviation of 12, the calculated confidence interval might look something like (72.37, 77.63). This means that we are 90% confident that the true population mean lies between approximately 72.37 and 77.63 based on the given sample data. The margin of error is half the width of the interval, and it represents the maximum likely difference between the sample mean and the true population mean.
```

```python
#Q6.

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parameters
df = 10  # Degrees of freedom
x = np.linspace(0, 30, 1000)  # Range of chi-square values

# Chi-square distribution
chi_square_dist = stats.chi2.pdf(x, df)

# Create the plot
fig, ax = plt.subplots()
ax.plot(x, chi_square_dist, 'b-', label=f'Chi-Square (df={df})')

# Shade the area corresponding to a chi-square statistic of 15
x_fill = np.linspace(0, 15, 1000)
y_fill = stats.chi2.pdf(x_fill, df)
ax.fill_between(x_fill, y_fill, color='lightgray', alpha=0.5, label='Area for stat = 15')

# Labels and title
ax.set_xlabel('Chi-Square Value')
ax.set_ylabel('Probability Density')
ax.set_title('Chi-Square Distribution with Shaded Area')
ax.legend()

plt.show()
```
